# Plotting Graphs for Training and Evaluation

```python
#import libraries and dependancies
import gym
import time
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
DDPG_test_avg = np.load("DDPG_test_avg.npy")
DQN_test_avg = np.load("DQN12:11:2021_17:32:58_test_avg.npy")
NFQ_test_avg = np.load("NFQAgent12:11:2021_16:32:10_test_avg.npy")
QLearn_Func_Approx_test_avg = np.load("QLearning_FuncApprox12:10:2021_07:21:49_test_avg.npy")
VPG_test_avg = np.load("VPG_test_avg.npy")
PPO_test_avg = np.load("PPO12:11:2021_17:37:37_test_avg.npy")
Tabular_Q_Learning_train_avg = np.load("Tabular_Q_Learning_train_avg.npy")
Random_Actions_train_avg = np.load("Random_Actions_train_avg.npy")
```

```python
DDPG_train_avg = np.load("DDPG_train_avg.npy")
DQN_train_avg = np.load("DQN12:11:2021_17:32:58_train_avg.npy")
NFQ_train_avg = np.load("NFQAgent12:11:2021_16:32:10_train_avg.npy")
QLearn_Func_Approx_train_avg = np.load("QLearning_FuncApprox12:10:2021_07:21:49_train_avg.npy")
VPG_train_avg = np.load("VPG_train_avg.npy")
PPO_train_avg = np.load("PPO12:11:2021_17:37:37_train_avg.npy")
Tabular_Q_Learning_train_avg = np.load("Tabular_Q_Learning_train_avg.npy")
Random_Actions_train_avg = np.load("Random_Actions_train_avg.npy")
```
